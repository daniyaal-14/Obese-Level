```python
import pandas as pd
```

```python
dt = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")
```

```python
dt.head()
```

```python
dt.iloc[:10, :11]
```

```python
dt.info()
```

```python
dt["NObeyesdad"].value_counts()
```

```python
dt["Gender"].replace({"Male": 0, "Female": 1}, inplace = True)
dt["family_history_with_overweight"].replace({"yes":1, "no":0}, inplace = True)
dt["FAVC"].replace({"yes":1, "no":0}, inplace = True)
dt["SMOKE"].replace({"yes":1, "no":0}, inplace = True)
dt["SCC"].replace({"yes":1, "no":0}, inplace = True)
```

```python
dt.head()
```

```python
dt.rename(columns = {"Gender": "Female"}, inplace = True)
#1 means person is female 0 means person is male
```

```python
from sklearn.preprocessing import OrdinalEncoder
val = [["no", "Sometimes", "Frequently", "Always"]]
```

```python
dt["CAEC"] = OrdinalEncoder(categories = val).fit_transform(dt[["CAEC"]])
dt["CALC"] = OrdinalEncoder(categories = val).fit_transform(dt[["CALC"]])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(), ["MTRANS"]),
                                    remainder = "passthrough",
                                    verbose_feature_names_out = False)

onehot_df = col_trans.fit_transform(dt)

dt = pd.DataFrame(onehot_df, columns = col_trans.get_feature_names_out())
```

```python
dt["NObeyesdad"].unique()
```

```python
x = dt.drop(["NObeyesdad"],axis=1)
y = dt["NObeyesdad"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 14)
```

```python
tree = DecisionTreeClassifier(max_depth = 4, random_state = 14)
```

```python
tree.fit(x_train, y_train)
```

```python
tree.score(x_test, y_test)
```
